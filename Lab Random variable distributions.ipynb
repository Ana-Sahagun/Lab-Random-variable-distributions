```python
var_numericas=df.select_dtypes(include=['int', 'float'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear grÃ¡ficos de distribuciÃ³n para las variables numÃ©ricas
sns.pairplot(var_numericas, diag_kind='kde', kind='scatter')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Aplicar transformaciÃ³n logarÃ­tmica a las variables numÃ©ricas
var_numericas_log = np.log(var_numericas)

# Crear grÃ¡ficos de distribuciÃ³n para las variables numÃ©ricas transformadas
sns.pairplot(var_numericas_log, diag_kind='kde', kind='scatter')
plt.show()
```
